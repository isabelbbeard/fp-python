```python
path_to_chromedriver = '/Users/isabelbeard/Downloads/chromedriver' # change path as needed
browser = webdriver.Chrome(executable_path = path_to_chromedriver)
url = 'https://www.skincarisma.com/search'
browser.get(url)
```

```python
%%time

#create empty dataframe
products = pd.DataFrame(columns=['id', 'product', 'brand', 'ratingsCount', 'rating', 'price', 'categories'])
ingredients = pd.DataFrame(columns = ['id', 'ingredient', 'function', 'notes', 'ewg', 'cir'])
index = 0

for link in links:
    url = link
    browser.get(url)

    page = browser.find_element_by_id('main-product')
    
    #get product name
    try:
        product = page.find_element_by_class_name('card-title').text
    except:
        product = 'NaN'
    
    #get brand
    try:
        brand = page.find_element_by_class_name('card-subtitle').text
    except:
        brand = 'NaN'
    
    #Get number of ratings
    try:
        rating_group = page.find_element_by_class_name('flex-between')
        num_ratings = rating_group.find_element_by_css_selector('span').text
    except:
        num_ratings = 'NaN'
           
    #Get rating
    try:
        rating = page.find_element_by_class_name('h6').text
    except:
        rating = 'NaN'
           
    try:
        all_notes = []
        notes = page.find_elements_by_class_name('p-2')
        for note in notes:
            all_notes.append(note.text)
    except:
        all_notes = 'Nan'
    
    try:
        categories_div = page.find_elements_by_class_name('category-badge')
        category = []

        for categories in categories_div:
            category.append(categories.text)
        #    print(categories.text)

        category = list(filter(None, category))
    except:
        category = 'NaN'
    
    print('id:', index)
    print('product:', product)
    print('brand:', brand)
    print('number of ratings:', num_ratings)
    print('rating: ', rating)
    print('notes: ', all_notes)
    print('category: ', category)
    
    ##ingredient lists
    try:
        ingredient_link = page.find_element_by_link_text('Full Ingredient List')
        ingredient_link.click()
        print('clicked')
        ingredient_table = page.find_elements_by_class_name('ingredients-table')
        rows = ingredient_table.find_elements_by_tag_name('tr')
        for row in rows:
            cells = row.find_elements_by_tag_name('td')
            print(cells.text)
    except:
        print('No ingredients')
    index = index + 1

    time.sleep(5)
    
    
```

```python
#Get all liks for possible searches
search_links = []
    
for i in range(0, 985):
    i = i + 1
    search_link = 'https://www.skincarisma.com/search?page=' + str(i)
    search_links.append(search_link)

print(search_links)
```

```python
##get product links
#%%time

links = []

for i in range(0, len(search_links)):
    
    print('page:', i + 1)
    url = search_links[i]
    browser.get(url)
    table = browser.find_element_by_css_selector('ul.mt-2')
    items = table.find_elements_by_tag_name('li')
    
    for item in items:
        brand_product = item.find_elements_by_class_name('d-inline-block')
        product = brand_product[1].text

        link = item.find_element_by_partial_link_text(product)
        link = link.get_attribute("href")
        links.append(link)
    try: 
        next = browser.find_element_by_class_name('next_page ')
        next.click()
    
    except:
        print('done')

print(links)
```
